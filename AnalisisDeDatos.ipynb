```python
#VENTA POR ORIGEN(MONEDA)
Total_online=df2.loc[df2["Origen"]=="Online"]
VO=sum(Total_online["Total"])
print("Ventas Online --->",VO)

Total_fisica=df2.loc[df2["Origen"]=="Tienda Física"]
VF=sum(Total_fisica["Total"])
print("Ventas Fisica --->",VF)

x=["Online","Fisica"]
y=[VO,VF]

fig, ax=plt.subplots()
ax.bar(x,y,linewidth=5)
plt.show()
```

```python
Totales_origen=df2[["Total","Origen"]]
Totales_origen2= Totales_origen.groupby(["Origen"]).sum()

fig, ax=plt.subplots()
valores= Totales_origen2["Total"]
etiquetas= df2["Origen"].unique()
ax.bar(x=etiquetas,height=valores)
```

```python
df2.info()
```

```python
cantidad=df2[df2["Origen"]=="Online"]
a=cantidad["Origen"].count()

cantidad2=df2[df2["Origen"]=="Tienda Física"]
b=cantidad2["Origen"].count()

print("Cantidad de ventas Online --->",a)
print("Cantidad de ventas Fisica --->",b)

```

```python
cantidado=df2[df2["Origen"].isin(["Online"])]
to=cantidado["Total"].count()
print(to)

cantidado2=df2[df2["Origen"].isin(["Tienda Física"])]
tf=cantidado2["Total"].count()
print(tf)

#VALORES A UTILIZAR
Titulos=df2["Origen"].unique()
datos=[to,tf]
ax.pie(datos,labels=Titulos,autopct="%1.1f%%")
plt.show()
```

```python
dftotales=df2[["Producto","Total"]]
dftotales2=dftotales.groupby(["Producto"]).sum()
dftotales2

```

```python
df22=df2.groupby(["Producto"])["Total"].transform("sum").unique()
fig, ax=plt.subplots()
ax.pie(df22,labels=df22,autopct="%1.1f%%")
plt.show()
```
